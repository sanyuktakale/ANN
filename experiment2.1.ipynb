```python
import numpy as np
import matplotlib.pyplot as plt

# Hypothesis function
def predict(x, m, b):
    return m * x + b

# Compute cost function
def compute_cost(x, y, m, b):
    n = len(y)
    cost = (1/(2*n)) * np.sum((predict(x, m, b) - y) ** 2)
    return cost

# Gradient Descent algorithm
def gradient_descent(x, y, m, b, learning_rate, iterations):
    n = len(y)
    cost_history = []

    for _ in range(iterations):
        y_pred = predict(x, m, b)

        # Compute gradients
        dm = (-2/n) * np.sum(x * (y - y_pred))
        db = (-2/n) * np.sum(y - y_pred)

        # Update parameters
        m -= learning_rate * dm
        b -= learning_rate * db

        cost = compute_cost(x, y, m, b)
        cost_history.append(cost)

    return m, b, cost_history

# Sample dataset: House size (sq ft) vs Price ($1000s)
x = np.array([600, 800, 1000, 1200, 1400, 1600, 1800, 2000, 2200, 2400])
y = np.array([150, 200, 250, 275, 300, 320, 350, 400, 420, 450])

# Initialize parameters
m = 0  # Slope
b = 0  # Intercept
learning_rate = 0.0000001
iterations = 10000

# Run gradient descent
m, b, cost_history = gradient_descent(x, y, m, b, learning_rate, iterations)

# Predictions using trained model
predictions = predict(x, m, b)

# Plot data and regression line
plt.scatter(x, y, color='red', label='Actual Prices')
plt.plot(x, predictions, color='blue', label='Predicted Prices')
plt.xlabel("House Size (sq ft)")
plt.ylabel("Price ($1000s)")
plt.title("House Price Prediction using Gradient Descent")
plt.legend()
plt.show()

# Display final parameters
print(f"Final equation: y = {m:.4f}x + {b:.4f}")
```
